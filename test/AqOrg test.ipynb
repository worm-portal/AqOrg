### Readying the aqueous organic estimator

````python
aoe = AqOrg.AqOrganicEstimator()
````

### Load a custom table of group contribution values

AqOrg has a built-in table of group contribution values. If you want to replace it with your own custom table, use ```load_group_data```.

````python
aoe.load_group_data("custom_groups.csv")
aoe.df_gc # display the custom data
````

### Single compound estimation

````python
molecule = "triglycerol"
````

````python
aoe.display_molecule(molecule, save=False)
````

````python
aoe.test_group_match(molecule, show=False)
````

````python
aoe.estimate(molecule, show=False)
````

### Multi-compound estimation

````python
input_name = "alcohol_demo.csv"
output_name = "alcohol_demo_out.csv"
csv_out_name = "alcohol_demo_est.csv"
OBIGT_filename = "alcohol_OBIGT.csv"
````

````python
aoe.estimate(input_name=input_name,
             output_name=output_name,
             csv_out_name=csv_out_name,
             ig_method="Joback",
             OBIGT_filename=OBIGT_filename,
             show=False)
````
